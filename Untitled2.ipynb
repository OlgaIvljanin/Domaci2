```
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense 
import numpy as np

np.random.seed(3)

# broj klasa vina
classifications = 3

# ucitavanje dataset-a
dataset = np.loadtxt('vina.csv', delimiter=",")

#delimo dataset u dva skupa, jedna za testiranje i jedan za treniranje neuralne mreze

X = dataset[:,1:14]
Y = dataset[:,0:1]
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.66, random_state=5)

y_train = keras.utils.to_categorical(y_train-1, classifications)
y_test = keras.utils.to_categorical(y_test-1, classifications)


# kreiranje modela
model = Sequential()
model.add(Dense(10, input_dim=13, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(2, activation='relu'))
model.add(Dense(classifications, activation='softmax'))

#kompilacije i fitovanje modela
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
model.fit(x_train, y_train, batch_size=15, epochs=250, validation_data=(x_test, y_test))
predictionsX = model.predict(x_train)

#ispisivanje predikcije i ocekivanih vrednosti
print(predictionsX)
print(y_train)
```
